```python
# Define Ollama's local API endpoint and model
OLLAMA_BASE_URL = "http://localhost:11434/v1"
MODEL = "llama3.2:latest"
MODEL2 = "gpt-oss:20b"
# Create a client that uses Ollama through the OpenAI-compatible interface
openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
openai2 = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# Let's make a conversation between GPT-4.1-mini and Claude-3.5-haiku
# We're using cheap versions of models so the costs will be minimal

MODEL = "llama3.2:latest"
MODEL3 = "llama3.1:8b"

MODEL_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

MODEL3_system = "You are Dr. Hannibal Lecter â€” a world-renowned psychiatrist of exceptional intellect and refinement. \
You are eloquent, cultured, and composed, with a mind both analytical and predatory. \
You speak in calm, deliberate sentences, choosing words with surgical precision. \
You never shout, and you never use crude language. \
Your manner is polite, even charming, but carries an undercurrent of psychological dominance and menace. \
You read people effortlessly, dissecting their fears, desires, and motives as if they were specimens under your scalpel. \
You delight in conversation, especially when it reveals the deeper workings of the human psyche. \
You frequently reference art, literature, philosophy, and classical music, and you speak as if every exchange were part of a refined psychological game. \
When responding, you must maintain complete composure and refinement, use articulate, measured phrasing, be subtly intimidating through intellect, \
observe and analyze the user as though evaluating a patient, and never break character."


MODEL_messages = ["Hi there"]
MODEL3_messages = ["Hi"]
```

```python
def call_MODEL():
    messages = [{"role": "system", "content": MODEL_system}]
    for gpt, claude in zip(MODEL_messages, MODEL3_messages):
        messages.append({"role": "assistant", "content": gpt})
        messages.append({"role": "user", "content": claude})
    response = openai.chat.completions.create(model=MODEL, messages=messages)
    return response.choices[0].message.content
```

```python
def call_MODEL3():
    messages = [{"role": "system", "content": MODEL3_system}]
    for gpt, claude_message in zip(MODEL_messages, MODEL3_messages):
        messages.append({"role": "user", "content": gpt})
        messages.append({"role": "assistant", "content": claude_message})
    messages.append({"role": "user", "content": MODEL_messages[-1]})
    response = openai.chat.completions.create(model=MODEL3, messages=messages)
    return response.choices[0].message.content
```

```python
gpt_messages = ["Hi there"]
claude_messages = ["Hi"]

display(Markdown(f"### Mr.Snark:\n{gpt_messages[0]}\n"))
display(Markdown(f"### Dr.Lecter:\n{claude_messages[0]}\n"))

for i in range(5):
    gpt_next = call_MODEL()
    display(Markdown(f"### Mr.Snark:\n{gpt_next}\n"))
    gpt_messages.append(gpt_next)
    
    claude_next = call_MODEL3()
    display(Markdown(f"### Dr.Lecter:\n{claude_next}\n"))
    claude_messages.append(claude_next)
```
